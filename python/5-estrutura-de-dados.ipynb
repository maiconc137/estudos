# Estrutura de Dados
Estruturas de dados são formas de organizar, gerenciar e armazenar dados de maneira eficiente.
Em Python, as principais estruturas de dados incluem listas, tuplas, dicionários e conjuntos.


## Intrudução a listas
Listas em Python são coleções ordenadas e mutáveis de itens. 
Elas podem armazenar diferentes tipos de dados, como inteiros, strings, ou até mesmo outras listas.

```python
# Criando uma lista
fruits = ["apple", "banana", "cherry"]
```

## Manipulando Listas 
Listas em Python são estruturas de dados que armazenam uma coleção de itens.

```python
# Acessando elementos da lista (indexação começa em 0)
print("Primeira fruta:", fruits[0])

# Modificando um item na lista
fruits[1] = "blueberry"
print("Lista modificada:", fruits) 

# Verificando o comprimento da lista
print("Número de frutas:", len(fruits))
```

## Funções dentro de Listas
Python oferece diversas funções e métodos embutidos para manipular listas, como adicionar, remover, 
ordenar e modificar elementos.

```python
# Criando uma lista de números
numbers = [5, 2, 9, 1, 5, 6]

# Função len(): Retorna o número de elementos na lista
print("Comprimento da lista:", len(numbers))  # Saída: 6

# Função sorted(): Retorna uma nova lista ordenada
sorted_numbers = sorted(numbers)
print("Lista ordenada:", sorted_numbers)  # Saída: [1, 2, 5, 5, 6, 9]

# Método append(): Adiciona um novo elemento ao final da lista
numbers.append(10)
print("Lista após adicionar 10:", numbers)  # Saída: [5, 2, 9, 1, 5, 6, 10]

# Método remove(): Remove a primeira ocorrência de um elemento específico
numbers.remove(5)
print("Lista após remover o primeiro 5:", numbers)  # Saída: [2, 9, 1, 5, 6, 10]

# Método pop(): Remove e retorna o último elemento da lista (ou o elemento no índice especificado)
last_item = numbers.pop()
print("Elemento removido:", last_item)  # Saída: 10
print("Lista após pop():", numbers)  # Saída: [2, 9, 1, 5, 6]

# Método reverse(): Inverte a ordem dos elementos na lista
numbers.reverse()
print("Lista invertida:", numbers)  # Saída: [6, 5, 1, 9, 2]

# Método extend(): Adiciona todos os elementos de outra lista
numbers.extend([3, 4])
print("Lista após extend:", numbers)  # Saída: [6, 5, 1, 9, 2, 3, 4]
```

## Concatenando listas
Em Python, você pode concatenar (juntar) duas ou mais listas utilizando o operador de adição (+) 
ou o método extend(), que adiciona elementos de uma lista à outra.

```python
# Exemplo: Concatenando listas com o operador +
list1 = [1, 2, 3]
list2 = [4, 5, 6]
concatenated_list = list1 + list2

print("Lista concatenada com +:", concatenated_list)  # Saída: [1, 2, 3, 4, 5, 6]

# Exemplo: Usando o método extend() para adicionar os elementos de list2 em list1
list1.extend(list2)
print("Lista após usar extend():", list1)  # Saída: [1, 2, 3, 4, 5, 6]
```

## Extraindo variáveis de listas
Em Python, você pode atribuir elementos de uma lista a variáveis individuais usando a técnica de desempacotamento (unpacking).
Isso é útil quando você sabe exatamente quantos elementos há na lista.

```python
# Exemplo: Desempacotando uma lista
fruits = ["apple", "banana", "cherry"]

# Atribuindo cada elemento da lista a uma variável
fruit1, fruit2, fruit3 = fruits

# Exibindo as variáveis
print("Fruta 1:", fruit1)  # Saída: apple
print("Fruta 2:", fruit2)  # Saída: banana
print("Fruta 3:", fruit3)  # Saída: cherry

# Exemplo: Desempacotamento parcial com o operador "*"
# Podemos usar o operador * para capturar o restante dos valores
numbers = [1, 2, 3, 4, 5]

first, second, *rest = numbers
print("Primeiro número:", first)  # Saída: 1
print("Segundo número:", second)  # Saída: 2
print("Restante:", rest)  # Saída: [3, 4, 5]
```

## Looping dentro de uma lista
Para percorrer (loop) os elementos de uma lista em Python, podemos usar a estrutura de repetição `for`.
Isso permite acessar e manipular cada item da lista.

```python
# Exemplo: Loop simples sobre uma lista de números
numbers = [10, 20, 30, 40, 50]

for number in numbers:
    print("Número:", number)

# Exemplo: Loop com strings
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(f"Fruta: {fruit.capitalize()}")  # O método capitalize() torna a primeira letra maiúscula

# Exemplo: Loop com índice e elemento usando enumerate()
for index, number in enumerate(numbers):
    print(f"Índice {index} tem o número {number}")
```

## Verificando itens em uma lista
Para verificar se um item está presente em uma lista, podemos usar o operador `in`, que retorna True ou False.

```python
# Exemplo: Verificando a presença de um item
fruits = ["apple", "banana", "cherry"]

if "banana" in fruits:
    print("Banana está na lista de frutas.")
else:
    print("Banana não está na lista.")

# Exemplo: Verificando a ausência de um item
if "orange" not in fruits:
    print("Orange não está na lista de frutas.")

# Exemplo: Verificando múltiplos itens com um loop
items_to_check = ["banana", "orange", "apple"]

for item in items_to_check:
    if item in fruits:
        print(f"{item.capitalize()} está na lista.")
    else:
        print(f"{item.capitalize()} não está na lista.")
```

## Agregando duas listas com Zip
A função zip() em Python combina dois ou mais iteráveis (como listas) em pares de tuplas.
Isso permite que você agregue elementos correspondentes de várias listas.

```python
# Exemplo: Agregando duas listas
names = ["Alice", "Bob", "Charlie"]
ages = [25, 30, 35]

# Usando zip() para combinar as listas
combined = zip(names, ages)

# Convertendo o zip em uma lista e exibindo
combined_list = list(combined)
print("Lista combinada:", combined_list)  # Saída: [('Alice', 25), ('Bob', 30), ('Charlie', 35)]

# Exemplo: Iterando sobre duas listas usando zip
for name, age in zip(names, ages):
    print(f"{name} tem {age} anos.")
```

## Input em uma Lista

Podemos usar a função input() para permitir que o usuário insira valores em uma lista.
Esses valores podem ser coletados e adicionados à lista dinamicamente.

```python
# Exemplo: Coletando 3 entradas do usuário e adicionando em uma lista
user_inputs = []

for i in range(3):
    item = input(f"Digite o item {i + 1}: ")
    user_inputs.append(item)

# Exibindo a lista final com os itens inseridos
print("Lista de itens inseridos:", user_inputs)

# frutas_usuario = input('fruta: ')

# fruta_lista = frutas_usuario.split(', ')

# print(fruta_lista)
```

## Entendendo sobre Tuples

Tuples são coleções imutáveis de itens em Python. Uma vez criadas, suas estruturas não podem ser alteradas.
Elas são semelhantes às listas, mas são definidas com parênteses em vez de colchetes e são úteis para armazenar dados que não devem ser modificados. (seu uso é mais eficiente por consumir menos memória para ser armazenada)

```python
# Exemplo: Criando uma tuple
my_tuple = (1, 2, 3, "apple", "banana")

# Acessando elementos da tuple
print("Primeiro elemento:", my_tuple[0])  # Saída: 1
print("Último elemento:", my_tuple[-1])  # Saída: banana

# Tentando modificar um elemento (irá gerar um erro)
# my_tuple[1] = 20  # Descomente esta linha para ver o erro

# Tuples podem conter outros tipos de dados, inclusive outras tuples
nested_tuple = (1, (2, 3), "apple")
print("Elemento aninhado:", nested_tuple[1][0])  # Saída: 2

# Desempacotamento de tuples
a, b, c, *rest = my_tuple
print("a:", a)  # Saída: 1
print("b:", b)  # Saída: 2
print("c:", c)  # Saída: 3
print("rest:", rest)  # Saída: ['apple', 'banana']
```

## Trabalhando com Arrays
Em Python, arrays são similares a listas, mas são mais eficientes para operações matemáticas e científicas.
Eles são disponibilizados pelo módulo `array`, que fornece uma maneira de armazenar dados de maneira mais compacta e eficiente.
Além disso, bibliotecas como NumPy oferecem arrays mais poderosos e flexíveis para cálculos avançados.


```python
import array

# Exemplo: Criando um array com o módulo array
# O tipo 'i' indica que o array contém inteiros
int_array = array.array('i', [1, 2, 3, 4, 5])

# Acessando e modificando elementos do array
print("Primeiro elemento:", int_array[0])  # Saída: 1

# Modificando um elemento
int_array[1] = 10
print("Array após modificação:", int_array)  # Saída: array('i', [1, 10, 3, 4, 5])
```

## Criando sets
Em Python, um set (conjunto) é uma coleção não ordenada de elementos únicos e imutáveis. Sets são úteis quando você precisa armazenar itens sem duplicatas e realizar operações matemáticas como união, interseção e diferença.

Você pode criar um set usando chaves {} ou a função set(). No entanto, para criar um set vazio, você deve usar set() em vez de {}, pois {} cria um dicionário vazio.

```python
# Criando um set com elementos
frutas = {"maçã", "banana", "laranja"}
print(frutas)

# Tentando adicionar um elemento duplicado
frutas.add("maçã")
print(frutas)  # O set permanece o mesmo porque "maçã" já existe

# Criando um set vazio e adicionando elementos
conjunto_vazio = set()
conjunto_vazio.add(1)
conjunto_vazio.add(2)
print(conjunto_vazio)

# Operações com sets
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

# União
uniao = set_a.union(set_b)
print("União:", uniao)

# Interseção
intersecao = set_a.intersection(set_b)
print("Interseção:", intersecao)

# Diferença
diferenca = set_a.difference(set_b)
print("Diferença:", diferenca)
```

## Funções em sets
Em Python, os sets (conjuntos) possuem vários métodos (funções) embutidos que permitem manipular e realizar operações com conjuntos de forma eficiente. Essas funções facilitam a adição e remoção de elementos, bem como a execução de operações matemáticas como união, interseção e diferença.

```python
# Criando um set inicial
numeros = {1, 2, 3}
print("Set inicial:", numeros)

# Adicionando um elemento
numeros.add(4)
print("Após adicionar 4:", numeros)

# Removendo um elemento existente
numeros.remove(2)
print("Após remover 2:", numeros)

# Tentando remover um elemento inexistente com remove() - gera erro
# numeros.remove(5)  # Descomente esta linha para ver o erro

# Removendo um elemento inexistente com discard() - não gera erro
numeros.discard(5)
print("Após descartar 5 (inexistente):", numeros)

# Removendo um elemento arbitrário com pop()
elemento_removido = numeros.pop()
print(f"Elemento removido com pop(): {elemento_removido}")
print("Set após pop():", numeros)

# Limpando o set
numeros.clear()
print("Após clear():", numeros)

# Operações com sets
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

# União
uniao = set_a.union(set_b)
print("União de A e B:", uniao)

# Interseção
intersecao = set_a.intersection(set_b)
print("Interseção de A e B:", intersecao)

# Diferença
diferenca = set_a.difference(set_b)
print("Diferença de A e B:", diferenca)

# Diferença simétrica
dif_simetrica = set_a.symmetric_difference(set_b)
print("Diferença simétrica de A e B:", dif_simetrica)

# Verificando subconjunto e superconjunto
set_c = {1, 2}
print("C é subconjunto de A?", set_c.issubset(set_a))
print("A é superconjunto de C?", set_a.issuperset(set_c))

# Copiando um set
set_d = set_a.copy()
print("Cópia de A:", set_d)
```

## Sets com strings
Em Python, você pode criar sets (conjuntos) que contêm strings como elementos. Isso é útil para manipular coleções de palavras ou caracteres, realizar operações de conjuntos e eliminar duplicatas. Strings são objetos imutáveis e podem ser usados como elementos de um set. Você pode aplicar todas as operações e métodos de sets a conjuntos de strings da mesma forma que faz com outros tipos de dados imutáveis.

```python
# Criando sets de strings
alunos_turma_a = {"Ana", "Bruno", "Carlos", "Diana"}
alunos_turma_b = {"Carlos", "Diana", "Eduardo", "Fernanda"}

# Imprimindo os sets
print("Alunos da Turma A:", alunos_turma_a)
print("Alunos da Turma B:", alunos_turma_b)
```

## Introdução a dicionário
Em Python, um dicionário é uma estrutura de dados que armazena pares de chave-valor. As chaves são únicas dentro de um dicionário e são usadas para armazenar e acessar valores associados. Os dicionários são mutáveis, o que significa que você pode alterar, adicionar ou remover pares de chave-valor após a criação do dicionário.


```python
# Criando um dicionário com informações de um aluno
aluno = {
    'nome': 'Maria',
    'idade': 20,
    'curso': 'Engenharia',
    'notas': [9.5, 8.7, 10.0]
}

# Acessando valores no dicionário
print("Nome:", aluno['nome'])
print("Idade:", aluno['idade'])
print("Curso:", aluno['curso'])
print("Notas:", aluno['notas'])

# Adicionando um novo par chave-valor
aluno['universidade'] = 'Universidade Federal'
print("Universidade:", aluno['universidade'])

# Atualizando um valor existente
aluno['idade'] = 21
print("Idade atualizada:", aluno['idade'])

# Removendo um par chave-valor
del aluno['curso']
print("Dicionário após remover 'curso':", aluno)

# Iterando sobre as chaves do dicionário
print("Informações do aluno:")
for chave in aluno:
    print(chave, ":", aluno[chave])

# Usando o método get() para acessar valores
endereco = aluno.get('endereco', 'Endereço não disponível')
print("Endereço:", endereco)

# Verificando se uma chave existe no dicionário
if 'notas' in aluno:
    print("Notas do aluno:", aluno['notas'])
else:
    print("Notas não encontradas.")

# Obtendo apenas as chaves ou valores
chaves = aluno.keys()
valores = aluno.values()
print("Chaves:", chaves)
print("Valores:", valores)

# Criando um dicionário vazio e adicionando pares chave-valor
dicionario_vazio = {}
dicionario_vazio['a'] = 1
dicionario_vazio['b'] = 2
print("Dicionário vazio preenchido:", dicionario_vazio)
```
