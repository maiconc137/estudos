# Operadores

Em Python, os operadores são usados para realizar operações em variáveis e valores. Eles podem ser classificados em várias categorias


##Operações matemáticas em Python


```python
# Python suporta operações matemáticas básicas, como adição, subtração, multiplicação, divisão e muito mais.

# Adição (+)
a = 10
b = 5
sum_result = a + b
print(f"Addition: {a} + {b} = {sum_result}")  # 10 + 5 = 15

# Subtração (-)
sub_result = a - b
print(f"Subtraction: {a} - {b} = {sub_result}")  # 10 - 5 = 5

# Multiplicação (*)
mul_result = a * b
print(f"Multiplication: {a} * {b} = {mul_result}")  # 10 * 5 = 50

# Divisão (/)
div_result = a / b
print(f"Division: {a} / {b} = {div_result}")  # 10 / 5 = 2.0

# Divisão inteira (//) - retorna o quociente da divisão sem considerar o resto
div_int_result = a // b
print(f"Integer Division: {a} // {b} = {div_int_result}")  # 10 // 5 = 2

# Módulo (%) - retorna o resto da divisão
mod_result = a % b
print(f"Modulus: {a} % {b} = {mod_result}")  # 10 % 5 = 0

# Exponenciação (**) - eleva um número à potência de outro
exp_result = a ** b
print(f"Exponentiation: {a} ** {b} = {exp_result}")  # 10 ** 5 = 100000

# Também é possível combinar operações matemáticas:
combined_result = (a + b) * (a - b)
print(f"Combined Operations: ({a} + {b}) * ({a} - {b}) = {combined_result}")  # (10 + 5) * (10 - 5) = 75

# Além dessas operações básicas, Python também suporta operações com números de ponto flutuante e possui uma biblioteca
# matemática, chamada 'math', que oferece funções avançadas, como raiz quadrada, seno, cosseno, etc.
```

## Operadores de comparação

```python
# Operadores de comparação são usados para comparar dois valores e retornam True ou False.

a = 10
b = 5
c = 10

# Igualdade (==)
print(f"{a} == {b}: {a == b}")  # False, porque 10 não é igual a 5
print(f"{a} == {c}: {a == c}")  # True, porque 10 é igual a 10

# Diferença (!=)
print(f"{a} != {b}: {a != b}")  # True, porque 10 é diferente de 5
print(f"{a} != {c}: {a != c}")  # False, porque 10 não é diferente de 10

# Maior que (>)
print(f"{a} > {b}: {a > b}")  # True, porque 10 é maior que 5
print(f"{a} > {c}: {a > c}")  # False, porque 10 não é maior que 10

# Menor que (<)
print(f"{a} < {b}: {a < b}")  # False, porque 10 não é menor que 5
print(f"{a} < {c}: {a < c}")  # False, porque 10 não é menor que 10

# Maior ou igual (>=)
print(f"{a} >= {b}: {a >= b}")  # True, porque 10 é maior que 5
print(f"{a} >= {c}: {a >= c}")  # True, porque 10 é igual a 10

# Menor ou igual (<=)
print(f"{a} <= {b}: {a <= b}")  # False, porque 10 não é menor ou igual a 5
print(f"{a} <= {c}: {a <= c}")  # True, porque 10 é igual a 10
```

## Operadores de atribuição

```python
# Operadores de atribuição são usados para atribuir valores a variáveis.

# Atribuição simples (=)
x = 10  # Atribui o valor 10 à variável x
print(f"x = 10: {x}")

# Atribuição com adição (+=)
x += 5  # Equivale a x = x + 5
print(f"x += 5: {x}")  # 10 + 5 = 15

# Atribuição com subtração (-=)
x -= 3  # Equivale a x = x - 3
print(f"x -= 3: {x}")  # 15 - 3 = 12

# Atribuição com multiplicação (*=)
x *= 2  # Equivale a x = x * 2
print(f"x *= 2: {x}")  # 12 * 2 = 24

# Atribuição com divisão (/=)
x /= 4  # Equivale a x = x / 4
print(f"x /= 4: {x}")  # 24 / 4 = 6.0

# Atribuição com divisão inteira (//=)
x //= 2  # Equivale a x = x // 2
print(f"x //= 2: {x}")  # 6.0 // 2 = 3.0

# Atribuição com módulo (%=)
x %= 2  # Equivale a x = x % 2
print(f"x %= 2: {x}")  # 3.0 % 2 = 1.0

# Atribuição com exponenciação (**=)
x **= 3  # Equivale a x = x ** 3
print(f"x **= 3: {x}")  # 1.0 ** 3 = 1.0
```
