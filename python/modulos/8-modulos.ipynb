# Módulos em Python

Módulos são arquivos Python que contêm definições e implementações de funções, variáveis e classes. Eles permitem organizar o código em diferentes arquivos, facilitando a reutilização e manutenção. Um módulo pode ser importado e utilizado em qualquer outro arquivo Python.

Existem três tipos principais de módulos:
1. **Módulos padrão**: Fazem parte da biblioteca padrão do Python, como `math`, `datetime`, `os`, entre outros.
2. **Módulos criados pelo usuário**: Arquivos `.py` que o desenvolvedor cria para organizar seu código.
3. **Módulos de terceiros**: Pacotes que podem ser instalados via `pip` e utilizados em seu projeto, como `pandas`, `numpy`, etc.

### Como usar um módulo:
Para usar um módulo, utilizamos a palavra-chave `import`. O Python irá procurar o arquivo do módulo e importar seu conteúdo para ser utilizado no código atual.

Exemplo:
```python
import math

# Utilizando funções do módulo math
resultado = math.sqrt(16)
print(resultado)  # Saída: 4.0
