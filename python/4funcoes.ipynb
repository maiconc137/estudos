# Funções em Python

Uma função é um bloco de código que pode ser reutilizado. Ela é definida com a palavra-chave 'def' e pode receber parâmetros e retornar valores.

## De Functions a Libraries
Em Python, podemos organizar múltiplas funções em módulos e pacotes, formando uma biblioteca. 
Isso facilita a reutilização e a manutenção do código.

```python
# Exemplo: Criando uma função simples
def greet(name):
    """
    Esta função recebe um nome e retorna uma saudação personalizada.
    """
    return f"Hello, {name}!"

# Exemplo de como agrupar funções em um arquivo para criar um módulo (por exemplo, 'mymodule.py'):
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

# Esses módulos podem ser importados e usados em outros arquivos Python:
# Exemplo de uso em outro arquivo:
# from mymodule import add, subtract

# result = add(10, 5)
# print(result)  # Saída: 15
```

## Como funciona uma função
Uma função em Python é um bloco de código reutilizável que executa uma tarefa específica. 
Ela é definida usando a palavra-chave `def`, pode receber parâmetros (dados de entrada), 
executar operações e retornar um valor.

```python
# Exemplo: função que soma dois números
def add(a, b):
    """
    Esta função recebe dois números (a e b) como parâmetros e retorna a soma deles.
    """
    result = a + b  # Executa a soma
    return result  # Retorna o resultado

# Chamando a função com diferentes valores
sum1 = add(5, 3)
sum2 = add(10, 20)

print(f"A soma de 5 e 3 é: {sum1}")  # Saída: A soma de 5 e 3 é: 8
print(f"A soma de 10 e 20 é: {sum2}")  # Saída: A soma de 10 e 20 é: 30
```

## Criando uma função de soma

```python
# Definindo a função 'sum_numbers' que recebe dois parâmetros e retorna a soma
def sum_numbers(a, b):
    """
    Esta função recebe dois números (a e b) e retorna a soma deles.
    """
    return a + b  # Retorna a soma dos dois números

# Chamando a função com diferentes valores
result1 = sum_numbers(5, 10)
result2 = sum_numbers(3, 7)

# Exibindo os resultados
print(f"A soma de 5 e 10 é: {result1}")  # Saída: A soma de 5 e 10 é: 15
print(f"A soma de 3 e 7 é: {result2}")   # Saída: A soma de 3 e 7 é: 10
```

## Parâmetros e Argumentos em uma função
Parâmetros são variáveis que aparecem na definição da função. 
Argumentos são os valores reais que você passa para a função quando a chama.

```python
# Exemplo: função que calcula a área de um retângulo
def calculate_area(length, width):
    """
    Esta função recebe dois parâmetros (length e width) e retorna a área de um retângulo.
    """
    return length * width  # Calcula e retorna a área

# Chamando a função com argumentos específicos
area1 = calculate_area(5, 3)  # Aqui 5 e 3 são os argumentos
area2 = calculate_area(7, 2)

print(f"A área do retângulo com comprimento 5 e largura 3 é: {area1}")  # Saída: 15
print(f"A área do retângulo com comprimento 7 e largura 2 é: {area2}")  # Saída: 14
```

## Argumentos Default e Non-default
Argumentos default têm um valor padrão atribuído a eles quando a função é definida.
Se o usuário não fornecer um valor para esse argumento ao chamar a função, o valor padrão será usado.

```python
# Exemplo: função que calcula o salário de um funcionário com um bônus opcional
def calculate_salary(base_salary, bonus=500):
    """
    Esta função calcula o salário total, com um bônus opcional.
    Se o bônus não for especificado, o valor padrão de 500 será usado.
    """
    return base_salary + bonus

# Chamando a função com e sem o argumento de bônus
salary1 = calculate_salary(3000)  # Aqui usamos o valor default de bônus (500)
salary2 = calculate_salary(3000, 1000)  # Aqui fornecemos um bônus personalizado (1000)

print(f"Salário com bônus padrão: {salary1}")  # Saída: 3500
print(f"Salário com bônus personalizado: {salary2}")  # Saída: 4000
```

## Print ou Return em Funções
O 'print' exibe um valor na tela, mas não armazena ou retorna o valor para uso posterior.
O 'return' devolve um valor que pode ser armazenado, manipulado ou reutilizado em outros lugares do código.

```python
# Exemplo 1: Função com print
def print_square(num):
    """
    Esta função exibe o quadrado do número, mas não retorna nada.
    """
    print(f"O quadrado de {num} é {num ** 2}")

# Exemplo 2: Função com return
def return_square(num):
    """
    Esta função calcula o quadrado do número e o retorna para que possa ser usado.
    """
    return num ** 2

# Usando a função com print
print_square(4)  # Saída: O quadrado de 4 é 16

# Usando a função com return
square = return_square(4)  # O valor 16 é retornado e armazenado na variável 'square'
print(f"O quadrado armazenado é: {square}")  # Saída: O quadrado armazenado é: 16
```

## Vários argumentos xargs com números
O *args permite que você passe um número variável de argumentos para uma função.
Esses argumentos são recebidos como uma tupla dentro da função.

```python
# Exemplo: Função que soma uma quantidade variável de números
def sum_all(*args):
    """
    Esta função aceita múltiplos números como argumento e retorna a soma deles.
    """
    return sum(args)  # A função sum() soma todos os elementos da tupla args

# Chamando a função com diferentes quantidades de números
result1 = sum_all(1, 2, 3)         # Soma de 1 + 2 + 3
result2 = sum_all(5, 10, 15, 20)   # Soma de 5 + 10 + 15 + 20
result3 = sum_all(10, 20)          # Soma de 10 + 20

print(f"A soma de 1, 2 e 3 é: {result1}")           # Saída: 6
print(f"A soma de 5, 10, 15 e 20 é: {result2}")     # Saída: 50
print(f"A soma de 10 e 20 é: {result3}")            # Saída: 30
```

## Iimportando módulo
Em Python, módulos são arquivos que contêm definições e declarações que podem ser reutilizadas em outros programas. 
Para usar funções e classes de um módulo, é necessário importá-lo primeiro.

```python
# Exemplo: Usando o módulo math para calcular o fatorial de um número
import math  # Importando o módulo math, que contém várias funções matemáticas

# Função para calcular o fatorial de um número
def factorial_of_number(n):
    """
    Esta função calcula e retorna o fatorial de n.
    """
    return math.factorial(n)  # Utiliza a função factorial() do módulo math

# Chamando a função para calcular o fatorial de 6
number = 6
result = factorial_of_number(number)

# Exibindo o resultado
print(f"O fatorial de {number} é: {result}")  # Saída: O fatorial de 6 é: 720
```
