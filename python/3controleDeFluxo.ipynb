# Controle de fluxo

Controle de fluxo em Python permite direcionar a execução do código com base em condições e repetições. Com ele, é possível criar programas que tomam decisões e realizam tarefas repetidas vezes, dependendo de variáveis e condições que mudam durante a execução. Isso é essencial para desenvolver lógicas dinâmicas e interativas em qualquer aplicação.


## If, Else
As declarações `if` e `else` permitem que você controle o fluxo de execução do programa com base em condições.


```python
# Exemplo básico de if:
age = 18

if age >= 18:
    print("You are an adult.")  # Esta linha será executada porque a condição é verdadeira

# Exemplo de if-else:
age = 16

if age >= 18:
    print("You are an adult.")
else:
    print("You are not an adult.")  # Esta linha será executada porque a condição if é falsa

# Exemplo de if-elif-else:
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")  # Esta linha será executada porque a condição if é falsa e a condição elif é verdadeira
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: F")

# Usando operadores lógicos (and, or) nas condições:
temperature = 25

if temperature > 20 and temperature < 30:
    print("The weather is warm.")  # Esta linha será executada porque ambas as condições são verdadeiras
else:
    print("The weather is not warm.")

# if 20 < temperature < 30:
#     print("The weather is warm.")  # Esta linha será executada porque ambas as condições são verdadeiras
# else:
#     print("The weather is not warm.")

# Você também pode usar condições mais complexas:
number = 7

if number % 2 == 0:
    print(f"{number} is even.")  # Verifica se o número é par
else:
    print(f"{number} is odd.")   # Caso contrário, o número é ímpar
```

## Operadores lógicos
Os operadores lógicos em Python são usados para combinar várias condições em uma única expressão.
Os operadores lógicos mais comuns são: `and`, `or` e `not`.

```python
# Exemplo de and:
# O operador 'and' retorna True se ambas as condições forem verdadeiras
age = 25
has_id = True

if age >= 18 and has_id:
    print("You are allowed to enter.")  # Esta linha será executada porque ambas as condições são verdadeiras

# Exemplo de or:
# O operador 'or' retorna True se pelo menos uma das condições for verdadeira
is_student = False
is_senior = True

if is_student or is_senior:
    print("You are eligible for a discount.")  # Esta linha será executada porque pelo menos uma condição é verdadeira

# Exemplo de not:
# O operador 'not' inverte o valor lógico de uma condição
is_raining = False

if not is_raining:
    print("You don't need an umbrella.")  # Esta linha será executada porque not False é True

# Operadores lógicos podem ser combinados para formar expressões mais complexas:
temperature = 22
humidity = 75

if temperature > 20 and humidity > 70 or not is_raining:
    print("The weather is good for a walk.")  # Esta linha será executada porque a expressão completa é verdadeira

# Exemplo envolvendo comparações:
x = 10
y = 5
z = 15

if x > y and x < z:
    print(f"{x} is between {y} and {z}.")  # Esta linha será executada porque ambas as condições são verdadeiras
```

## Múltiplus operadores de comparação
Operadores de comparação são usados para comparar valores e retornar um resultado booleano (`True` ou `False`).

```python
# Igualdade (==): Retorna True se os valores comparados forem iguais
a = 10
b = 20
print(f"{a} == {b}: {a == b}")  # False

# Diferença (!=): Retorna True se os valores comparados forem diferentes
print(f"{a} != {b}: {a != b}")  # True

# Maior que (>): Retorna True se o valor da esquerda for maior que o valor da direita
print(f"{a} > {b}: {a > b}")  # False

# Menor que (<): Retorna True se o valor da esquerda for menor que o valor da direita
print(f"{a} < {b}: {a < b}")  # True

# Maior ou igual a (>=): Retorna True se o valor da esquerda for maior ou igual ao valor da direita
print(f"{a} >= {b}: {a >= b}")  # False

# Menor ou igual a (<=): Retorna True se o valor da esquerda for menor ou igual ao valor da direita
print(f"{a} <= {b}: {a <= b}")  # True

# Comparações encadeadas:
x = 5
y = 10
z = 15

# Verifica se x está entre y e z
print(f"{x} < {y} < {z}: {x < y < z}")  # True

# Verifica se x é menor que y e y é menor que z
print(f"{x} < {y} and {y} < {z}: {x < y and y < z}")  # True
```

## For Loop - Utilizando números
O `loop for` em Python é usado para iterar sobre uma sequência de valores, como uma lista ou uma faixa de números.

```python
# Exemplo básico usando range() para iterar sobre uma faixa de números
for i in range(5):  # Itera sobre os números de 0 a 4
    print(i)  # Imprime cada número

# Exemplo usando range() com início, fim e passo
for i in range(1, 10, 2):  # Itera sobre os números de 1 a 9 com passo 2
    print(i)  # Imprime 1, 3, 5, 7, 9

# Exemplo de loop for para calcular a soma de números
total = 0
for i in range(1, 6):  # Itera sobre os números de 1 a 5
    total += i  # Adiciona cada número ao total
print(f"Sum of numbers from 1 to 5: {total}")  # Imprime a soma

# Exemplo de loop for para criar uma lista de quadrados de números
squares = []
for i in range(5):  # Itera sobre os números de 0 a 4
    squares.append(i ** 2)  # Adiciona o quadrado de i à lista
print(f"Squares of numbers from 0 to 4: {squares}")  # Imprime a lista de quadrados

# Exemplo de loop for com números negativos
for i in range(-3, 1):  # Itera sobre os números de -3 a 0
    print(i)  # Imprime -3, -2, -1, 0
```

## For Loop - Utilizando Strings
O `loop for` em Python pode ser usado para iterar sobre caracteres individuais de uma string.

```python
# Exemplo básico de iteração sobre uma string
message = "Hello"

for char in message:
    print(char)  # Imprime cada caractere da string

# Exemplo de contagem de caracteres em uma string
count = 0
for char in message:
    count += 1
print(f"Total number of characters in '{message}': {count}")  # Imprime a contagem de caracteres

# Exemplo de encontrar e imprimir apenas as vogais em uma string
vowels = "aeiou"
for char in message:
    if char in vowels:
        print(f"Vowel found: {char}")  # Imprime as vogais encontradas

# Exemplo de construir uma nova string com caracteres em maiúsculas
uppercase_message = ""
for char in message:
    uppercase_message += char.upper()  # Converte cada caractere para maiúscula e adiciona à nova string
print(f"Uppercase version: {uppercase_message}")  # Imprime a string em maiúsculas

# Exemplo de inverter uma string usando um loop for
reversed_message = ""
for char in message:
    reversed_message = char + reversed_message  # Adiciona cada caractere ao início da nova string
print(f"Reversed string: {reversed_message}")  # Imprime a string invertida
```

## For Loop - Utilizando If e Else
O `loop for` pode ser combinado com `if` e `else` para criar lógica condicional dentro da iteração.

```python
# Exemplo básico: verificar se cada número em uma lista é par ou ímpar
numbers = [1, 2, 3, 4, 5, 6]

for num in numbers:
    if num % 2 == 0:  # Verifica se o número é divisível por 2 (ou seja, se é par)
        print(f"{num} is even.")
    else:
        print(f"{num} is odd.")  # Caso contrário, o número é ímpar

# Exemplo de categorização de uma lista de strings
words = ["apple", "banana", "cherry", "date"]

for word in words:
    if len(word) > 5:  # Verifica se a palavra tem mais de 5 caracteres
        print(f"'{word}' is a long word.")
    else:
        print(f"'{word}' is a short word.")  # Caso contrário, é uma palavra curta

# Exemplo: verificar se um caractere em uma string é uma vogal ou consoante
text = "hello"

for char in text:
    if char in "aeiou":  # Verifica se o caractere é uma vogal
        print(f"'{char}' is a vowel.")
    else:
        print(f"'{char}' is a consonant.")  # Caso contrário, é uma consoante

# Exemplo: filtrar números positivos e negativos de uma lista
values = [-10, 20, -30, 40, 0]

for value in values:
    if value > 0:  # Verifica se o número é positivo
        print(f"{value} is positive.")
    elif value < 0:  # Verifica se o número é negativo
        print(f"{value} is negative.")
    else:
        print(f"{value} is zero.")  # Caso contrário, o número é zero
```

## Nested loops
Loops aninhados (`nested loops`) ocorrem quando um loop está dentro de outro.

```python
# Exemplo básico: criar uma tabela de multiplicação (1 a 3)
for i in range(1, 4):  # Loop externo, percorre os números de 1 a 3
    for j in range(1, 4):  # Loop interno, também percorre os números de 1 a 3
        print(f"{i} * {j} = {i * j}")  # Imprime o resultado da multiplicação
    print("---")  # Separador entre as iterações do loop externo

# Exemplo: criar um padrão de asteriscos em forma de triângulo
rows = 5
for i in range(1, rows + 1):  # Loop externo controla o número de linhas
    for j in range(i):  # Loop interno controla o número de asteriscos por linha
        print("*", end="")  # Imprime asteriscos na mesma linha
    print()  # Move para a próxima linha após imprimir todos os asteriscos da linha atual

# Exemplo: combinar duas listas usando loops aninhados
colors = ["Red", "Green", "Blue"]
objects = ["Ball", "Pen", "Book"]

for color in colors:  # Loop externo percorre a lista de cores
    for obj in objects:  # Loop interno percorre a lista de objetos
        print(f"{color} {obj}")  # Combina a cor com o objeto

# Exemplo: comparar elementos de duas listas
list1 = [1, 2, 3]
list2 = [3, 2, 1]

for x in list1:  # Loop externo percorre os elementos de list1
    for y in list2:  # Loop interno percorre os elementos de list2
        if x == y:  # Verifica se os elementos são iguais
            print(f"{x} is equal to {y}")
```

## For Loop - Separando Strings
Usando `for loop` para separar e processar palavras em uma string

```python
# Exemplo básico: separar uma string em palavras
sentence = "Python is a powerful language"

# O método split() separa a string com base nos espaços por padrão
words = sentence.split()

for word in words:
    print(f"Word: {word}")  # Imprime cada palavra separadamente

# Exemplo: separar uma string por outro delimitador (vírgula)
data = "apple,banana,cherry"

# O método split(",") separa a string com base na vírgula
fruits = data.split(",")

for fruit in fruits:
    print(f"Fruit: {fruit}")  # Imprime cada item da lista separada por vírgula

# Exemplo: processar cada caractere de uma string
text = "Hello"

for char in text:
    print(f"Character: {char}")  # Imprime cada caractere da string

# Exemplo: contar quantas vezes cada palavra aparece em uma string
sentence2 = "apple banana apple cherry banana apple"
word_count = {}

# Separa a string em palavras
words2 = sentence2.split()

for word in words2:
    if word in word_count:
        word_count[word] += 1  # Incrementa o contador se a palavra já estiver no dicionário
    else:
        word_count[word] = 1  # Adiciona a palavra ao dicionário com contador 1

print("Word counts:", word_count)  # Imprime a contagem de cada palavra

# Exemplo: inverter as palavras de uma string
sentence3 = "Learning Python is fun"
words3 = sentence3.split()

reversed_sentence = ""
for word in words3:
    reversed_sentence = word + " " + reversed_sentence  # Adiciona a palavra no início da nova string

print(f"Reversed sentence: {reversed_sentence.strip()}")  # Imprime a frase invertida
```

## For Loop - Criando um retângulo

```python
# Definindo o número de linhas (altura) e colunas (largura) do retângulo
height = 4  # Altura do retângulo
width = 20   # Largura do retângulo

# Usando dois loops for aninhados para criar o retângulo
for i in range(height):  # Loop externo para controlar as linhas
    for j in range(width):  # Loop interno para controlar as colunas
        print("*", end="")  # Imprime asteriscos sem mudar de linha
    print()  # Move para a próxima linha após imprimir os asteriscos de uma linha
```

## Conhecendo o While Loop
O loop while continua a execução enquanto uma condição é verdadeira.

```python
# Exemplo simples: imprimir números de 1 a 3

count = 1  # Inicializa a variável de controle

while count <= 3:  # O loop continua enquanto count for menor ou igual a 3
    print(count)  # Imprime o valor atual de count
    count += 1  # Incrementa count a cada iteração

print("Loop finalizado!")  # Mensagem final após o loop
```

## Operador Ternário
O operador ternário é uma forma curta de escrever uma condição `if-else`.

```python
# Exemplo básico: verificar se um número é par ou ímpar
number = 5

# Sintaxe: valor_if_true if condição else valor_if_false
result = "Even" if number % 2 == 0 else "Odd"

print(result)  # Imprime 'Odd' porque 5 não é divisível por 2
```

##  Diferenças entre For Loop e While Loop

O `for loop` é usado quando se sabe o número de iterações ou se está iterando sobre uma sequência.
O `while loop` é utilizado quando não se sabe quantas iterações serão necessárias, mas há uma condição que deve ser satisfeita.

```python
# Exemplo de For Loop: iterando uma sequência de números
for i in range(1, 6):  # Itera de 1 a 5
    print(f"For Loop iteration: {i}")

# Exemplo de While Loop: rodando até que a condição seja falsa
count = 1
while count <= 5:  # O loop continua enquanto a condição for verdadeira
    print(f"While Loop iteration: {count}")
    count += 1
```

## Criando condições com while loop
O `while loop` continua executando o bloco de código enquanto uma condição for verdadeira.
Ele é útil para criar loops que dependem de uma condição que pode mudar dentro do loop.

```python
# Exemplo: Solicitar que o usuário adivinhe um número até acertar
secret_number = 7
guess = None  # Inicializamos a variável

while guess != secret_number:  # O loop continua até que o usuário adivinhe corretamente
    guess = int(input("Adivinhe o número (1 a 10): "))
    
    if guess < secret_number:
        print("Tente um número maior.")
    elif guess > secret_number:
        print("Tente um número menor.")
    else:
        print("Parabéns, você acertou!")


# Alternativa usando break
# while True:  # O loop continua indefinidamente até o break ser acionado
#     guess = int(input("Adivinhe o número (1 a 10): "))
    
#     if guess < secret_number:
#         print("Tente um número maior.")
#     elif guess > secret_number:
#         print("Tente um número menor.")
#     else:
#         print("Parabéns, você acertou!")
#         break  # O break encerra o loop quando o número correto é adivinhado
```
